```

# funcion a calcular: sen(2x), centrado en  x = 4


from sympy import Symbol, diff
import math 
import numpy as np


a = (math.pi)/4     ## valor central de la serie
x_ = 1              ## punto hasta donde se predice la serie de f(x)
h = x_ -a           ## tamaño de paso
grado = 5           ## 6 terminos 


x = Symbol('x')
funcion = cos(x)


F = [ funcion,  ]


for i in range(1, 7, 1):    ## hallar hasta la sexta derivada de la funcion
    der_ = diff(F[i-1], x)
    F.append(der_)
    #print('type: ', type(der_))


print(F)
```

```


F_2 = [[1, math.cos], [-1, math.sin], [-1, math.cos], [1, math.sin], [1, math.cos], [-1, math.sin], [-1, math.cos] ]


## evaluar la serie de Taylor y el error de truncamiento
suma = 0

listaTi = [ F_2[i][0] * F_2[i][1](i) for i in range(0, grado+2, 1)]
#print(lista)

polis = []
errores = [ ]
for i in range(0, grado+1, 1):
  poli = (listaTi[i] * h **i) /(np.math.factorial(i) ) 
  polis.append(poli)
  suma    = suma + poli
  error   =  ( listaTi[i+1] * h **(i+1) ) /(np.math.factorial(i+1) )
  errores.append(error)

#print('suma: ', suma)
print('errores: ', errores)

```

# Paso 4: Generar tabla de valores X error

```
from tabulate import tabulate

print(len(listaTi))
print(len(errores))
TablaSerieTaylor = [ [str(i), listaTi[i], errores[i] ]  for i in range(0, grado+1, 1 )  ]

print(tabulate(TablaSerieTaylor, headers= [ 'Orden', 'Término', 'Error' ]))
```
