# Summas de Riemann: punto intermedio.
## Definición de suma de Riemann.
Las sumas de Riemman son una aproximacion para obtener el area bajo una curva entre dos puntos especificos. Esta aproximacion se realiza usando poligonos regulares que cubren el area objetivo, de los cuales se puede hallar su area con mayor facilidad.

##Suma de punto medio.
Son una aproximación de Riemann que toman la evaluacion de cada punto de los intervalos en donde este punto es la mitad del ancho de la figura geometrica. 
Se observa que esta aproximacion, respecto a la derecha y a la izquierda, es mas precisa debido a que intenta distribuir las areas no cubiertas con las areas sobrantes de la figura semejante proxima.

```

# ej.  B  ::  punto medio :: funcion = x**3
##   Ejecutar :: imprime todo el cuadro de una vez.


from sympy import integrate, init_printing
from sympy.abc import x
#init_printing(use_latex="mathjax")

# valor real: se halla con sympy: con la funcion integrate
integ_v_exacto = integrate( x**3, (x, lim_inf, lim_sup) )

# en una funcion obtiene las sumas de Riemman para todas las particiones 3, 10, 25, 100 que estan en la tupla

def todas_las_sumas_Rnn( lim_inf, lim_sup, n_particiones ) :
  global particiones_ , evaluacion_Particion
  particiones_ , evaluacion_Particion = {}, {}
  sumasRiemann = []
  for n_particion in n_particiones :
    append = []
    append.append(n_particion)
    delta            =  (lim_sup - lim_inf) / n_particion
    ###                |---  I N T E R M E D I A   !!   ---|
    valores_x_delta  = [ ((lim_inf + delta / 2) + (i * delta)) for i in range( 0, n_particion, 1) ] 
    particiones_[n_particion] = valores_x_delta
    eval_func_parti  =  [ (  valor  **3 )  for valor in valores_x_delta   ]
    evaluacion_Particion[n_particion]   = eval_func_parti
    suma = 0
    for altura in eval_func_parti   :   suma += delta * altura 
    append.append(suma)
    error            = ( (integ_v_exacto - suma) / integ_v_exacto ) * 100 
    append.append(error)
    sumasRiemann.append(append)
  return sumasRiemann


lim_sup = 2
lim_inf = 0
n_particiones = ( 3, 10, 25, 100 )


sumasRiemann = todas_las_sumas_Rnn( lim_inf, lim_sup, n_particiones )


from tabulate import tabulate
print(tabulate( sumasRiemann ,  headers=['Partición', 'Suma', 'Error'], tablefmt= 'francy_grid'))
print('')

```
