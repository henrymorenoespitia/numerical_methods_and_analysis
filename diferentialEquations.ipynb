# EDO
### 1, 2, 4 order Runge-Kutta methods

```python
import matplotlib.pyplot as plt
import numpy as np


def f(x, y):
  return (y*(x**2)) + y


# dy/dx = f(x, y ) en el intervalo [0, 100] con y(0) = 1 ; 0<= x <= 2
# h es el tamaño de paso


# First order: Runge - Kutta


def runge_Kutta(order, function, initialPoint, finalPoint, moveValue,  y0 ):
  x = [initialPoint]
  y = [y0]
  if order == '1':
    while x[-1]<finalPoint:
      y.append(y[-1]+function(x[-1], y[-1])*moveValue)
      x.append(x[-1]+moveValue)
  elif order == '2':
    while x[-1]<finalPoint:
      k1=function(x[-1], y[-1])
      k2=function(x[-1]+3/4*moveValue, y[-1]+3/4*k1*moveValue)
      y.append(y[-1]+(1/3*k1+2/3*k2)*moveValue)
      x.append(x[-1]+moveValue)
  elif order == '4':
    while x[-1]<finalPoint:
      k1=f(x[-1], y[-1])
      k2=f(x[-1]+0.5*moveValue, y[-1]+0.5*k1*moveValue)
      k3=f(x[-1]+0.5*moveValue, y[-1]+0.5*k2*moveValue)
      k4=f(x[-1]+moveValue, y[-1]+k3*moveValue)
      y.append(y[-1]+(moveValue/6)*(k1+2*k2+2*k3+k4))
      x.append(x[-1]+moveValue)
  return x, y


h=10
print(f"Tamaño de paso: {h}")
x1, y1 = runge_Kutta('1', f, 0, 2,h, 1)
x2, y2 = runge_Kutta('2', f, 0, 2,h, 1)
x4, y4 = runge_Kutta('2', f, 0, 2,h, 1)
x5=np.arange(0.001,2, 0.01)
y5=1/((-x5)+1) # solucion analitica de la ED


plt.figure(figsize=(16,10), dpi=80)
plt.plot(x1, y1, label="R-K 1 order")
plt.plot(x2, y2, label="R-K 2 order")
plt.plot(x4, y4, label="R-K 4 order")
plt.plot(x5, y5, label="Real solution")
plt.legend()
plt.show()

```
